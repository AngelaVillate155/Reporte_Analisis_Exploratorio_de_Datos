```python
# Bloque 1: Importar Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bloque 2: Crear dataset ficticio de empleados
np.random.seed(42)
n = 200

departamentos = ["Ventas", "Finanzas", "IT", "Producción", "RRHH"]
generos = ["Masculino", "Femenino"]
niveles_educ = ["Bachiller", "Técnico", "Profesional", "Posgrado"]

empleados = pd.DataFrame({
"EmpleadoID": range(1, n+1),
"Edad": np.random.randint(22, 60, n),
"Género": np.random.choice(generos, n),
"Departamento": np.random.choice(departamentos, n),
"Salario": np.random.randint(1500, 8000, n),
"Antigüedad": np.random.randint(0, 15, n),
"NivelEducacion": np.random.choice(niveles_educ, n),
"Satisfacción": np.random.randint(1, 11, n),
"Rotación": np.random.choice([0, 1], n, p=[0.7, 0.3])
})
```

```python
# Bloque 3. Pregunta 1: ¿Cuál es la distribución de edades?
# =====================================================
plt.figure(figsize=(8,5))
sns.histplot(empleados["Edad"], bins=15, kde=True)
plt.title("Distribución de Edades de Empleados")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()


print("➡ La mayoría de empleados se concentran entre los 25 y 40 años, con pocos mayores de 50.")
```

```python
# Bloque 4. Pregunta 2: ¿Existen diferencias salariales entre departamentos?
# ======================================================
plt.figure(figsize=(8,5))
sns.boxplot(x="Departamento", y="Salario", data=empleados)
plt.title("Salarios por Departamento")
plt.xticks(rotation=45)
plt.show()


print("➡ IT y Finanzas presentan salarios más altos, mientras que RRHH y Ventas tienden a salarios menores.")
```

```python
# Bloque 5. Pregunta 3: ¿Qué departamentos tienen mayor rotación?
# ======================================================
rotacion_dept = empleados.groupby("Departamento")["Rotación"].mean()*100
rotacion_dept.sort_values().plot(kind="bar", figsize=(8,5))
plt.title("Porcentaje de Rotación por Departamento")
plt.ylabel("% de Rotación")
plt.show()


print("➡ Producción y Ventas tienen los porcentajes de rotación más altos.")
```

```python
# Bloque 6. Pregunta 4: ¿Cómo influye la satisfacción en la rotación?
# ======================================================
plt.figure(figsize=(8,5))
sns.boxplot(x="Rotación", y="Satisfacción", data=empleados)
plt.title("Satisfacción vs Rotación")
plt.xticks([0,1],["Permanece","Renunció"])
plt.show()


print("➡ Los empleados que renunciaron tienen en promedio, niveles de satisfacción más bajos.")
```

```python
# Bloque 7. Pregunta 5: ¿Existe relación entre antigüedad y rotación?
# ======================================================
plt.figure(figsize=(8,5))
sns.histplot(data=empleados, x="Antigüedad", hue="Rotación", multiple="stack", bins=12)
plt.title("Antigüedad y Rotación")
plt.xlabel("Años en la Empresa")
plt.show()


print("➡ La mayor rotación se observa en empleados con menos de 3 años en la empresa.")
```
